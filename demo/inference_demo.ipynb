```python
from mmseg.apis import init_segmentor, inference_segmentor, show_result_pyplot
from mmseg.core.evaluation import get_palette
```

```python
config_file = '../configs/segformer/segformer_mit-b5_512x512_160k_ade20k.py'
checkpoint_file = '../checkpoints/segformer_mit-b5_512x512_160k_ade20k_20210726_145235-94cedf59.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_segmentor(config_file, checkpoint_file, device='cuda:0')
```

```python
import cv2
img = cv2.imread("nanyanglink.png")
# img = img[0:512,0:512]
img.shape
```

```python
#  test a single image
# img = 'demo.png'
result = inference_segmentor(model, img)
```

```python
# show the results
show_result_pyplot(model, img, result, get_palette('ade20k'))
```
